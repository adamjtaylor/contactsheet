```python
from tifffile import TiffFile
from matplotlib.image import imsave
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from skimage.util import montage

```

```python
svs_path = "/Users/ataylor/Downloads/HT 122 P1 S1H3 Vp1 U11.svs"
ome_path = "/Users/ataylor/Documents/projects/htan/htan_imaging_subset_2021-06-02/data/s3/HTA9_1_BA_L_ROI04.ome.tif"
```

```python
with TiffFile(svs_path) as tif:
    images = []
    channels = []
    for series in tif.series:
        i = series.levels[-1].asarray()
        images.append(i)
```

```python
plt.rcParams['figure.dpi'] = 500 # default for me was 75
```

```python
fig, axs = plt.subplots(len(images))
for index, image in enumerate(images):
    axs[index].imshow(image)
    axs[index].axis('off')

plt.savefig('test.png')
```
